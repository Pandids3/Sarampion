```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009])
y=np.array([535.3, 1064.1, 1437.9, 1922.2, 2253.6, 2638.4, 2866.1, 2996.2, 3133.7, 3311.6])
```

```python
def fx (x1, coef):
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p*x1**n
        n = n - 1
    return fx
```

```python
Year = 2010


for i in range(0,10):
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, Year)
    
    print(f"For polynomial degree {i} The prediction is {p}")
    x1 = np.linspace(2000, Year + 1, 1000)
    y1 = fx(x1, coef) # funcion
    plt.figure(figsize=[25, 10])
    plt.title("Cumulative deaths (in thousands) vs Year: Polynomial degree " + str(i), c = "red")
    plt.scatter(x,y,s=150,c='green')
    plt.plot(x1,y1,"--",linewidth=3,color='orange')
    plt.scatter(Year,p,s=300,c='red')
    plt.yticks(range(0,4500,200))
    plt.xticks(range(2000, 2012, 1))
    plt.grid("on")
    ax=plt.gca()
    ax.set_xlabel("$Year$")
    ax.set_ylabel("$Accumulate of deaths (In thousands)$")
    plt.savefig("img" + str(i)+".jpg", dpi=600)
    plt.show()
```

```python
Year = 2010
x = np.array([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009])
y = np.array([535.3, 1064.1, 1437.9, 1922.2, 2253.6, 2638.4, 2866.1, 2996.2, 3133.7, 3311.6])

degree = np.arange(0, 70 + 1, 1)
aproxi = np.array([])

closest_degrees = []
closest_diff = float('inf')

for i in degree:
    coef = np.polyfit(x, y, i)
    p = np.polyval(coef, Year)
    aproxi = np.append(aproxi, p)

    y_pred_vec = np.array([])
    for j in x:
        y_pred = np.polyval(coef, j)
        y_pred_vec = np.append(y_pred_vec, y_pred)

    MSE = (sum((y - y_pred_vec)**2)) / len(y)
    print(f"For polynomial degree of {i} the MSE is: {round(MSE, 1)}")
    
    diff = abs(p - 3450.9)
    if diff < closest_diff:
        closest_diff = diff
        closest_degrees = [i]
    elif diff == closest_diff:
        closest_degrees.append(i)

plt.figure(figsize=[20, 10])
plt.title("Degree of Polynomial vs Prediction")
plt.plot(degree, aproxi, "--", linewidth=2, color='red')

plt.xlabel("Polynomial Degree")
plt.ylabel("Prediction")
plt.xticks(np.arange(0, 71, step=2))
plt.grid(True)


closest_degrees_str = ", ".join([str(degree) for degree in closest_degrees])
description = f"The polynomial degrees that provide predictions closest to 3450.9 is: {closest_degrees_str}"


plt.text(0.5, 0.1, description, fontsize=15, transform=plt.gca().transAxes)
plt.show()
```
